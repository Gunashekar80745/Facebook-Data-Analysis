#### Explanatory Data Analysis

```python
#histogram
sns.set()
_ = plt.hist(fb_df['age'])
_ = plt.xlabel("no of people")
_ = plt.ylabel("age")
_ = plt.title("people on facebook")
plt.show()
```

```python
#histogram
sns.set()
_ = plt.hist(fb_df['tenure'])
_ = plt.xlabel("no of people")
_ = plt.ylabel("tenure")
_ = plt.title("people on facebook")
plt.show()
```

```python
#histogram
sns.set()
_ = plt.hist(fb_df['friend_count'])
_ = plt.xlabel("no of people")
_ = plt.ylabel("friend_count")
_ = plt.title("people on facebook")
plt.show()
```

```python
#histogram
sns.set()
_ = plt.hist(fb_df['dob_day'])
_ = plt.xlabel("no of people")
_ = plt.ylabel("age")
_ = plt.title("Day wise register")
plt.show()
```

```python
#histogram
sns.set()
_ = plt.hist(fb_df['dob_month'])
_ = plt.xlabel("no of people")
_ = plt.ylabel("age")
_ = plt.title("Month wise register")
plt.show()
```

```python
sns.boxplot(x ='friend_count', y ='likes', data = fb_df)
```

```python
friend_count=fb_df['friend_count']
likes=fb_df['likes']
plt.scatter(friend_count,likes)
plt.xlabel("friend_count")
plt.ylabel("likes ")
plt.title("friend_count vs likes")
plt.show()
```

```python
age=fb_df['age']
likes=fb_df['likes']
plt.scatter(age,likes)
plt.xlabel("friend_count")
plt.ylabel("likes ")
plt.title("friend_count vs likes")
plt.show()
```

```python
age=fb_df['age']
friendships_initiated=fb_df['friendships_initiated']
plt.scatter(age,friendships_initiated)
plt.xlabel("age")
plt.ylabel("likes ")
plt.title("age vs friendships_initiated")
plt.show()
```

```python
www_likes=fb_df['www_likes']
likes_received=fb_df['likes_received']
plt.scatter(age,likes_received)
plt.xlabel("www_likes")
plt.ylabel("likes_received ")
plt.title("www_likes vs likes_received")
plt.show()
```
